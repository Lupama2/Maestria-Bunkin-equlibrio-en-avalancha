# Gráficos de los test que lo requieran

```python
#Importo librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defino ctes
m = 9.11e-31*1e3 #[g]
K = 1.380649e-23*(1/1e-7) #constante de Boltzmann [ergio/K], obtenida de NIST
```

## Test 1
Grafico los resultados de llamar a f_maxwell_pura múltiples veces

```python
#Cargo los datos del archivo
file_test1 = "test1.txt"
datos = np.loadtxt(file_test1)

#Hago un histograma normalizado
plt.hist(datos, bins = 100, density = True)

#Ajusto una distribución de mawell-boltzmann
from scipy.stats import maxwell
param = maxwell.fit(datos)
epsilon = 1e-2
if np.abs(param[0] - 0)<epsilon and np.abs(param[1] - 1) < epsilon:
    print("Función maxwell pura")
else:
    print("Función maxwell no pura con parámetros: ", param)


#Grafico el ajuste
x = np.linspace(np.min(datos), np.max(datos), 100)
y = maxwell.pdf(x, *param)
plt.plot(x, y, 'r--')
plt.show()

```
